```python
# Importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Set display options
pd.set_option('display.max_columns', None)
plt.rc('axes', grid=True)

# Load Dataset (Replace with actual file path if local)
df = pd.read_csv('cust data.csv')

# Display basic info
df.info()
df.head()

# Check for missing values
df.isnull().sum()

# Fill missing values
df['gender'].fillna('Prefer not to say', inplace=True)
df['total_price'] = df['price_per_unit'] * df['quantity_x']
df['age'].replace(df['age'].max(), np.nan, inplace=True)  # Handling incorrect max age
df['age'].fillna(df['age'].median(), inplace=True)

# Convert date columns to datetime format
df['order_date'] = pd.to_datetime(df['order_date'])
df['delivery_date'] = pd.to_datetime(df['delivery_date'])

# Remove duplicates
df.drop_duplicates(inplace=True)

### **Exploratory Data Analysis (EDA)**
# Number of customers per gender
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='gender', palette='coolwarm')
plt.title('Customer Count by Gender')
plt.show()

# Number of orders per age group
df['age_group'] = df['age'].apply(lambda x: 'Youth' if x <= 24 else ('Seniors' if x > 64 else 'Adults'))
sns.countplot(data=df, x='age_group', palette='viridis')
plt.title('Orders by Age Group')
plt.show()

# Top-selling products
top_products = df.groupby('product_name')['quantity_x'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_products.values, y=top_products.index, palette='Blues_r')
plt.title('Top 10 Best-Selling Products')
plt.show()

# Monthly revenue trend
df['month'] = df['order_date'].dt.strftime('%Y-%m')
monthly_sales = df.groupby('month')['total_price'].sum().reset_index()
plt.figure(figsize=(10, 5))
sns.lineplot(data=monthly_sales, x='month', y='total_price', marker='o', color='green')
plt.xticks(rotation=45)
plt.title('Monthly Revenue Trend')
plt.show()

### **Customer Segmentation using RFM Analysis**
rfm_df = df.groupby('customer_id').agg({
    'order_date': 'max',
    'order_id': 'nunique',
    'total_price': 'sum'
}).reset_index()

# Compute Recency, Frequency, Monetary (RFM) scores
latest_order_date = df['order_date'].max()
rfm_df['recency'] = (latest_order_date - rfm_df['order_date']).dt.days
rfm_df.rename(columns={'order_id': 'frequency', 'total_price': 'monetary'}, inplace=True)
rfm_df.drop(columns=['order_date'], inplace=True)

# Display top customers based on RFM
rfm_df.sort_values(by='monetary', ascending=False).head(10)

### **Predicting Purchase Likelihood using Logistic Regression**
# Create target variable: customers who spent above median monetary value are "high-value" customers
median_monetary = rfm_df['monetary'].median()
rfm_df['high_value_customer'] = (rfm_df['monetary'] > median_monetary).astype(int)

# Selecting features and target variable
X = rfm_df[['recency', 'frequency', 'monetary']]
y = rfm_df['high_value_customer']

# Standardizing data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train Logistic Regression Model
model = LogisticRegression()
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f'Logistic Regression Accuracy: {accuracy:.2f}')
print('\nConfusion Matrix:\n', conf_matrix)
print('\nClassification Report:\n', classification_rep)

# Save the processed data
df.to_csv('processed_sales_data.csv', index=False)
rfm_df.to_csv('rfm_analysis.csv', index=False)
```
